```python
# CREATING BINARY-CLASS CLASSIFICATION DATASET { 2-Features} and { linearly-seperable}
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
plt.figure()
plt.title('SYNTHETICALLY PRODUCED BINATY-CLASS CLASSIFICATION')
x,y=make_classification(n_samples=100,n_features=2,n_informative=2,n_redundant=0,n_classes=2,random_state=0,n_clusters_per_class=1,flip_y = 0.1,class_sep = 0.5)
plt.scatter(x[:,0],x[:,1],c=y,s=30)
plt.xlabel('x1 --->')
plt.ylabel('x2 --->')
```

```python
# SPLIT THE DATA
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# LINEAR KERNEL / LINEAR-SEPERABLE-DATA-PRESENT
from sklearn.svm import SVC
svm=SVC(C=1,kernel='linear').fit(x_train,y_train)
print(svm.score(x_test,y_test)*100)
```
